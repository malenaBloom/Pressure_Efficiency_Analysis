```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

#import data
data = np.loadtxt('PressureUPDATE.txt',delimiter=',')
pressure = (data[:,0])
hits = (data[:,1])
uncertainty = np.sqrt(hits)
pressureArray = np.linspace(1,20,80)

#scatter plot
plt.plot(pressure, hits, ".", markersize = 10, label = "simulation data")
plt.errorbar(pressure, hits, yerr = uncertainty, linestyle = "", label = "error")
plt.xlabel("Pressure of He3 gas (bars)")
plt.ylabel("Hits Recorded (per run of 1000000)")
plt.title("Pressure & Efficiency of He3 Neutron Detector")

#fit
coeff, cov = np.polyfit(pressure, hits,2,cov=True)
yFit = coeff[0]*pressureArray**2 + coeff[1]*pressureArray + coeff[2]
plt.plot(pressureArray, yFit, label = "best fit")
plt.legend()



```
